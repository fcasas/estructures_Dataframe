## Exercici 2

```python
# Resumeix estadísticament les columnes d’interès
flights.describe()
```

```python
flights['ArrDelay'].describe()
```

```python
# dades faltants hi ha per columna
flights.isnull().sum()
```

```python
#Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)
flights['NetDelay'] = (flights.ArrDelay-flights.DepDelay)
# considerem que arriba tard quan ho fa amb més de 15 minuts de l'horari previst
flights['Delayed'] = (flights.ArrDelay >=15).astype(int)
# km/h
flights['AvgSpeed'] = ((flights.Distance * 1.609344) / (flights.AirTime / 60))

flights[['ArrDelay','DepDelay','NetDelay','Delayed', 'AvgSpeed']].head()
```

```python
# Taula de les aerolínies amb més endarreriments acumulats
flights.groupby('UniqueCarrier')['ArrDelay'].sum().sort_values(ascending=False)
```

```python
# Quins són els vols més llargs?
flights[['TailNum','Origin','Dest','AirTime']].sort_values(by='AirTime',ascending=False).head(10)
```

```python
# I els més endarrerits?
flights[['TailNum','Origin','Dest','NetDelay']].sort_values(by='NetDelay',ascending=False).head(10)
```
